```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

n_samples = 100000

# Beta Distribution (e.g., alpha=2, beta=5)
alpha, beta = 2, 5
beta_samples = np.random.beta(alpha, beta, n_samples)

# Beta-Binomial Distribution (e.g., n=10, alpha=2, beta=5)
n, a, b = 10, 2, 5
beta_binomial_samples = stats.betabinom.rvs(n, a, b, size=n_samples)

# Gamma Distribution (e.g., shape=2, scale=1)
shape, scale = 2, 1
gamma_samples = np.random.gamma(shape, scale, n_samples)

# Extended Negative Binomial Distribution (e.g., n=10, p=0.5, k=1)
n_ext, p = 10, 0.5
extended_negative_binomial_samples = stats.nbinom.rvs(n_ext, p, size=n_samples)

# Normal Distribution (e.g., mean=0, std=1)
mu, sigma = 0, 1
normal_samples = np.random.normal(mu, sigma, n_samples)

# Inverse Gamma Distribution (e.g., shape=3, scale=2)
shape, scale = 3, 2
inverse_gamma_samples = stats.invgamma.rvs(shape, scale=scale, size=n_samples)

# Student's T Distribution (e.g., degrees of freedom=5)
df = 5
student_t_samples = np.random.standard_t(df, n_samples)

# Plotting each distribution with KDE for smoothness
fig, axs = plt.subplots(3, 2, figsize=(12, 16))
fig.suptitle('Different Probability Distributions with KDE for smoothness')

# Beta
sns.histplot(beta_samples, bins=50, kde=True, ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Beta Distribution')

# Beta-Binomial
sns.histplot(beta_binomial_samples, bins=30, kde=False, ax=axs[0, 1], color='orange', discrete=True)
axs[0, 1].set_title('Beta-Binomial Distribution (Discrete)')

# Gamma
sns.histplot(gamma_samples, bins=50, kde=True, ax=axs[1, 0], color='green')
axs[1, 0].set_title('Gamma Distribution')

# Extended Negative Binomial
sns.histplot(extended_negative_binomial_samples, bins=30, kde=False, ax=axs[1, 1], color='red', discrete=True)
axs[1, 1].set_title('Extended Negative Binomial Distribution (Discrete)')

# Normal
sns.histplot(normal_samples, bins=50, kde=True, ax=axs[2, 0], color='purple')
axs[2, 0].set_title('Normal Distribution')

# Inverse Gamma
sns.histplot(inverse_gamma_samples, bins=50, kde=True, ax=axs[2, 1], color='brown')
axs[2, 1].set_title('Inverse Gamma Distribution')

plt.tight_layout
```
